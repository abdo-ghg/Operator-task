Question: Write a program that checks if a given year is a leap year. A year is leap if it is divisible by 4, EXCEPT if it is divisible by 100, UNLESS it is also divisible by 400.

```python
year = 2024

if (year % 400 == 0):
    print(f"{year} is a leap year")
elif (year % 100 == 0):
    print(f"{year} is not a leap year")
elif (year % 4 == 0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

```python
year = 2024

is_leap = (year % 4 == 0) and (year % 100 != 0 or year % 400 == 0)

print(f"{year} is {'a leap year' if is_leap else 'not a leap year'}")
```

```python
import calendar

year = 2024

print(f"{year} is {'a leap year' if calendar.isleap(year) else 'not a leap year'}")
```

Question: Assign a grade based on a score: 90-100 (A), 80-89 (B), 70-79 (C), 60-69 (D), <60 (F). Handle invalid inputs (>100 or <0).

```python
score = 97
if score > 100 or score < 0:
    print("invalid")
elif score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

```python
score = 88
grades = ['F', 'D', 'C', 'B', 'A']
if score > 100 or score < 0:
  print("invalid")
else:
  if(score >= 90):
    index = 4
  elif(score >= 80):
    index = 3
  elif(score >= 70):
    index = 2
  elif(score >= 60):
    index = 1
  else:
    index = 0

print(grades[index])
```

```python
score = 70
grade_map = {
    range(90, 101): 'A',
    range(80,  90): 'B',
    range(70,  80): 'C',
    range(60,  70): 'D',
    range(0,   60): 'F'
}

result = "invalid" if score < 0 or score > 100 else "F"

for r, g in grade_map.items():
    if score in r:
        result = g
        break

print(result)
```

Question: Calculate price. If amount > 1000, 10% discount. If amount > 500, 5% discount. Else 0%.

```python
amount = 600
discount = 0
```

```python
if amount > 1000:
    discount = 0.1
elif amount > 500:
    discount = 0.05
else:
    discount = 0

to_pay= amount - (amount * discount)
print(to_pay)
```

```python
discount = 0.10 if amount > 1000 else 0.05 if amount > 500 else 0.0
to_pay = amount * (1 - discount)
print(to_pay)
```

```python
thresholds = [0, 500, 1000]
rates     = [0.00, 0.05, 0.10]
idx = 0
for i, t in enumerate(thresholds):
    if amount > t:
        idx = i

discount = rates[idx]
to_pay = amount * (1 - discount)

print(to_pay)
```

```python
rules = [
    (1000, 0.10),
    ( 500, 0.05),
    (   0, 0.00)
]

discount = next((rate for limit, rate in rules if amount > limit), 0.0)

to_pay = amount * (1 - discount)
print(to_pay)
```

Question: Check if a string is a palindrome (reads the same forward and backward) using if.

```python
st = "ananana"
if st == st[::-1]:
    print("palindrome")
else:
    print("not palindrome")
```

```python
st = "ananana"

left = 0
right = len(st) - 1
is_palindrome = True

while left < right:
    if st[left] != st[right]:
        is_palindrome = False
        break
    left += 1
    right -= 1

print("palindrome" if is_palindrome else "not palindrome")
```

```python
st = "ananana"

reversed_st = ''.join(reversed(st))

if st == reversed_st:
    print("palindrome")
else:
    print("not palindrome")
```

```python
st = "ananana"

chars = list(st)
reversed_chars = chars[::-1]

if chars == reversed_chars:
    print("palindrome")
else:
    print("not palindrome")
```
